```python
#Renaming
to_rename = pd.read_csv('data/movie.csv', index_col = 'movie_title')
id_rnm = {'Spectre' : 'NYSM', 'Tangled': 'Good Will Hunting'} # {old_idx1 : new_idx1, old_idx2 : new_idx2}
renamed = to_rename.rename(index = id_rnm) #for columns --> columns = col_rnm
renamed
```

```python
#alternative renaming

idxs_rnm = to_rename.index
cols_rnm = to_rename.columns

idxs_rnm = idxs_rnm.tolist()
cols_rnm = cols_rnm.tolist()

idxs_rnm[0] = 'ABCD'
idxs_rnm[1] = 'EFGH'
cols_rnm[0] = 'QWER'
cols_rnm[1] = 'TYUI'

#print(idxs_rnm, cols_rnm)
to_rename.index = idxs_rnm
to_rename.columns = cols_rnm
print(to_rename)
```

```python
#Creating and deleting colmns from df
created = pd.read_csv('data/movie.csv')
created['has_seen'] = 0 #by assigning scalar value
created['all_likes'] = (created['director_facebook_likes'] + created['actor_1_facebook_likes'] + created['actor_2_facebook_likes'] + created['actor_3_facebook_likes']) #assigning values of other columns
created.head()
created['all_likes'].isnull().sum()
created['all_likes'] = created['all_likes'].fillna(0)
created['is_greater_total_cast_likes'] = created['cast_total_facebook_likes'] >= created['all_likes']
created['is_greater_total_cast_likes'].all() #all returns true over an axis

#dropping all_likes and including only actors
created = created.drop(['all_likes'], axis = 1)
created['all_actors_likes'] = (created['actor_1_facebook_likes'] + created['actor_2_facebook_likes'] + created['actor_3_facebook_likes']) #assigning values of other columns
created.head()
#created['all_actors_likes'].isnull().sum()
created['all_actors_likes'] = created['all_actors_likes'].fillna(0)
created['is_greater_total_cast_likes'] = created['cast_total_facebook_likes'] >= created['all_actors_likes']
created['is_greater_total_cast_likes'].all() #all returns boolean over an axis
created['%of_fb_likes_from_actor_likes'] = (created['is_greater_total_cast_likes'] / created['all_actors_likes'])
created.head()
created['%of_fb_likes_from_actor_likes'].min(), created['%of_fb_likes_from_actor_likes'].max()
```
