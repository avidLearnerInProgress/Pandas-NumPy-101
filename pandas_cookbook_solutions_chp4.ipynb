```python
#iloc and loc are indexing operators --> they are called indexers
#s[item] --> indexing operator
#s.loc[item] --> indexer

#iloc selects by integer location and works similar to Python lists
#loc selects only index label similar to Python dictionaries

import pandas as pd
import numpy as np
college_data = pd.read_csv('data/college.csv', index_col='INSTNM')
college_data.info()

city = college_data['CITY']
city.head()
```

```python
city.iloc[3]
city.iloc[10:30]
city.iloc[10:100:10]
```

```python
city.loc['Heritage Christian University']
```

```python
np.random.seed(1)
labels = list(np.random.choice(city.index, 4))
labels
```

```python
city.loc[labels]
```

```python
#When passing scalar value -> scalar is returned
#When passing list -> series is returned
city.iloc[[3]]
```

```python
#To get disjointed set of rows as DF
college_data.iloc[[60, 99, 32]]
```

```python
#To select rows and columns simultaneously
college_data.iloc[:3, :4] #selects first 3 rows and first 4 columns
college_data.iloc[:, [4,6]].head() #selects all rows of columns 4 and 6
```

```python
#To select rows and columns using both integer and label indexing
college_start = college_data.columns.get_loc('UGDS_WHITE')
college_end = college_data.columns.get_loc('UGDS_UNKN') + 1
college_start, college_end
college_data.iloc[:5, college_start:college_end]
```

```python
#.iat and .at selectors are faster then iloc and loc respectively

college_data = pd.read_csv('data/college.csv', index_col = 'INSTNM')
cn = 'Texas A & M University-College Station'
college_data.loc[cn, 'UGDS_WHITE'] #selects UGDS_WHITE from the university TAMI i.ie it just retrives value at x[a][b]

#this can be done as follows as well:
college_data.at[cn, 'UGDS_WHITE']
```

```python
%timeit college_data.at[cn, 'UGDS_WHITE']
```

```python
%timeit college_data.loc[cn, 'UGDS_WHITE'] #selects UGDS_WHITE from the university TAMI i.ie it just retrives value at x[a][b]
```

```python
#The above results show at is faster than loc


#Important:::
'''
>> df[item] # Where `df` is a DataFrame and item is some object
1. If item is a string then
    Find a column name that matches the item exactly
    Raise KeyError if there is no match
    Return the column as a Series
2. If item is a list of strings then
    Raise KeyError if one or more strings in item don't match columns
    Return a DataFrame with just the columns in the list
3. If item is a slice object then
    Works with either integer or string slices
    Raise KeyError if label from label slice is not in index
    Return all ROWS that are selected by the slice
4. If item is a list, Series or ndarray of booleans then
    Raise ValueError if length of item not equal to length of DataFrame
    Use the booleans to return only the rows with%wTrue in same location
'''
```
