```python
import pandas as pd 
import numpy as np

#filtering data in pandas can be done using boolean indexing
#selecting rows by providing boolean value for each row

movie = pd.read_csv('data/movie.csv', index_col = 'movie_title')
movie.head()
```

```python
#find if each movie duration is longer than 2 hrs

movie_ge_2_hrs = movie['duration'] > 120
movie_ge_2_hrs.head(10)
movie_ge_2_hrs.sum()
movie['duration'].dropna().ge(120).mean()
movie_ge_2_hrs.describe()
```

```python
movie_ge_2_hrs.value_counts(normalize=True)
```

```python
criteria1 = movie['imdb_score'] > 8
criteria2 = movie['content_rating'] == "PG-13"
criteria3 = ((movie.title_year < 2000) | (movie.title_year > 2009))
final_critera = criteria1 & criteria2 & criteria3
final_critera.head()
```

```python
crit1 = movie['imdb_score'] < 3 
crit2 = movie['content_rating'] == 'R'
crit3 = (movie['title_year'] >= 2000) | (movie['title_year'] <= 2010)
final_crit = crit1 & crit2 & crit3
```

```python
final_criteria_all = final_critera | final_crit
final_criteria_all.head()
movie[final_criteria_all].head()
```

```python
cols = ['imdb_score', 'content_rating', 'title_year']
movie_filtered = movie.loc[final_criteria_all, cols]
movie_filtered.head(10)
```

```python
#canonical example of boolean indexing
college = pd.read_csv('data/college.csv')
college[college['STABBR'] == 'TX'].head() #this selects all the college having stabbr as tx
```

```python
#To hae multiple boolean indexing we can use isin()

states = ['TX', 'CA', 'NY']
college[college['STABBR'].isin(states)]
```

```python
college2 = college.set_index('STABBR')
college2.index.is_monotonic

college3 = college2.sort_index()
college3.index.is_monotonic
```

```python
%timeit college[college['STABBR'] == 'TX']
%timeit college2.loc['TX']
%timeit college3.loc['TX']
```

```python
#Hence proved that the sorted index performs an order of magnitude faster than boolean selection.

college_uq = college.set_index('INSTNM')
college_uq.index.is_unique
college[college['INSTNM'] == 'Stanford University']
college_uq.loc['Stanford University']

%timeit college[college['INSTNM'] == 'Stanford University']
%timeit college_uq.loc['Stanford University']
```

```python
#lessons learned:
#When index is sorted Pandas uses binary search to find elements - this improves performance
#When index is unique - pandas uses hashtables which accounts for faster selection
```

```python
schlumberger = pd.read_csv('data/slb_stock.csv', index_col = 'Date', parse_dates = ['Date'])
schlumberger.head()
```

```python
#select closing price as series
slb_close = schlumberger['Close']
slb_summary = slb_close.describe(percentiles = [.1, .9])
slb_summary

#select all upper and lower closing prices
up = slb_summary.loc['90%']
lw = slb_summary.loc['10%']
criteria = (slb_close < lw) | (slb_close > up)
slb_top_bottom = slb_close[criteria]
slb_top_bottom.head()
```

```python
#here we plot the results
#so we have horizontal lines for our lw and up 
#the black plot are the closing prices 
#the resulting series is plotted in gray
import matplotlib.pyplot as plt
slb_close.plot(color = 'black', figsize = (12, 6))
slb_top_bottom.plot(marker = 'o', style = ' ', ms = 4, color = 'lightgray') #ms - marker size

xmin = criteria.index[0]
xmax = criteria.index[-1]
plt.hlines(y = [lw, up], xmin = xmin, xmax = xmax, color = 'black') #plot horizontal lines
plt.show()
```

```python
slb_close.plot(color = 'black', figsize = (12,6))
plt.hlines(y = [lw, up], xmin = xmin, xmax = xmax,color = 'lightgray')
plt.fill_between(x = criteria.index, y1 = lw, y2 = slb_close.values, color='black')
plt.fill_between(x = criteria.index, y1 = lw, y2 = slb_close.values, where = slb_close < lw, color = 'lightgray')
plt.fill_between(x = criteria.index, y1 = up, y2 = slb_close.values, where = slb_close > up, color = 'lightgray')
```

```python
#determining normality in stock market prices
amzz = pd.read_csv('data/amzn_stock.csv', index_col = 'Date', parse_dates = ['Date'])
amzz.head()
```

```python
#pct_change() - calculates percentage change between current and prior element
amazon_daily_return = amzz['Close'].pct_change()
amazon_daily_return = amazon_daily_return.dropna()
amazon_daily_return.hist(bins = 20)
```

```python
'''Normal distributions approximately follow the 68-95-99.7 rule--meaning that 68% 4.
of the data falls between 1 standard deviation of the mean, 95% between 2, and
99.7% between 3.'''
mean = amazon_daily_return.mean()
std = amazon_daily_return.std()

#z-score - number of sd away from the mean
abs_z_score = amazon_daily_return.sub(mean).abs().div(std)
print(abs_z_score)
```

```python
# % of returns taht are within 1,2 or 3 sd
pcts = [abs_z_score.lt(i).mean() for i in range(1, 4)]
print('{:.3f} fall within 1 standard deviation. '
'{:.3f} within 2 and {:.3f} within 3'.format(*pcts))
```

```python
#using the where method to preserve the size of df

movie = pd.read_csv('data/movie.csv', index_col='movie_title')
fb_likes = movie['actor_1_facebook_likes'].dropna()
fb_likes.head()
fb_likes.describe(percentiles=[.1, .25, .5, .75, .9]).astype(int)
fb_likes.hist()
```

```python
#The above visualisation shows that the df is highly skewed to the right with many observations more than an order of magnitude > median
criteria_high = fb_likes<20000
criteria_high.mean().round(2)
fb_likes.where(criteria_high).round(2)
```

```python
fb_likes_other = fb_likes.where(criteria_high, other = 20000).head()

#we can also put a ceil or floor on the missing values as follows
criteria_low = fb_likes > 300
#here the second chained where() uses the ceil
fb_likes_cap = fb_likes.where(criteria_high, other=20000).where(criteria_low, 300)
fb_likes_cap.head()
fb_likes_cap.hist()

#we have separate clip(), clip_upper() and clip_lower() which also perform the desired ceil and floor
```

```python
#demonstrating mask method
movie = pd.read_csv('data/movie.csv', index_col='movie_title')
c1 = movie['title_year'] >= 2010
c2 = movie['title_year'].isnull()
criteria = c1 | c2
movie.mask(criteria).dropna(how = 'all').head()
```

```python
#selecting and filtering using booleans indexing and loc, iloc

movie = pd.read_csv('data/movie.csv', index_col='movie_title')
c1 = movie['content_rating'] == 'G'
c2 = movie['imdb_score'] < 4
criteria = c1 & c2
movie_loc = movie.loc[criteria]
movie_loc.head()
```

```python
#we cannot use directly iloc on series of booleans
#So we need to use ndarray of booleans

#movie_iloc = movie.iloc[criteria] --> throws error
movie_iloc = movie.iloc[criteria.values]
movie_iloc.equals(movie_loc)
```
